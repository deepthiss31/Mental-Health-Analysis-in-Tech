```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
col = pd.read_csv('survey.csv')
```

```python
col.iloc[:,3].head()
```

```python
country_list=list(col.iloc[:,3])
```

```python
country_dict=dict()
for i in country_list:
    if i not in country_dict:
        country_dict[i]=1
    else:
        country_dict[i]=country_dict[i]+1

print(country_dict)
```

```python
activities=list(country_dict.keys())
slices=list(country_dict.values())
```

```python
print(slices)
```

```python
cols=['c','m','r','b','w','gray','silver','gold','tan','bisque','navy','plum','moccasin','k','maroon','coral','olive',
      'yellowgreen','lawngreen','lightgreen','g','skyblue','darkblue','slateblue','crimson','violet','y','pink','purple','aqua',
      'tomato','lime','cornsilk','yellow','indigo','thistle','orchid','lightpink','dimgray','snow','brown','oldlace',
      'palegreen','darkgreen','seagreen','aliceblue']
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=10
fig_size[1]=10
plt.rcParams["figure.figsize"] = fig_size 

plt.pie(slices,labels=None,colors=cols,startangle=90,shadow=True)
plt.legend(country_dict.items())
plt.title('country')
plt.show()
```
